# Read the data from json file

```python
import json

fd = open("records.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)
```

```python
records
```

# Purchase

```python
user_prod = str(input("Enter the Id of the product : "))
user_quant= int(input("Enter the quantity : "))

if(user_quant < int(records[user_prod ]["Quantity"])):
    print("---------------------")
    print("Product: " , records[user_prod]["name"])
    print("section:" , records[user_prod]["Section"])
    print("Price: " , records[user_prod]["Price"])
    

    print("---------------------")
    print("Total amount: " , int(user_quant)*int(records[user_prod]["Price"]) )
    print("******THANK YOU******")
    records[user_prod]["Quantity"] = int(records[user_prod]["Quantity"]) - int(user_quant)  
else:
    print("Sorry this much quantity is not in stock.")
    print("Availabled quantity is :" ,int(records[user_prod]["Quantity"])-1 )
if(records[user_prod ]["Quantity"] == "1"):
    records[user_prod ]["Stock"] = "NO"
```

# Update Inventory after purchase

```python
js = json.dumps(records)  #convert the dict into string

fd = open("records.json" , "w") #helps to store the record in our local system
fd.write(js)  #hence it is json file so we use .json extension
fd.close()
```
